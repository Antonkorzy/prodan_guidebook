```python
import pandas as pd
import numpy as np
#get data https://economics.mit.edu/faculty/angrist/data1/mhe/card
dat = pd.read_table('public.dat', sep="\s+", header=None)
dat.columns = ['SHEET'                                                                     
    ,'CHAINr '                                                                     
    ,'CO_OWNED'                                                                    
    ,'STATE'                                                                      
    ,'SOUTHJ'                                                                     
    ,'CENTRALJ'                                                                 
    ,'NORTHJ'                                                                      
    ,'PA1'                                                                         
    ,'PA2'                                                                         
    ,'SHORE'                                                                       
    ,'NCALLS'                                                                      
    ,'EMPFT'                                                                       
    ,'EMPPT'                                                                       
    ,'NMGRS'                                                                       
    ,'WAGE_ST'                                                                     
    ,'INCTIME'                                                                     
    ,'FIRSTINC'                                                                    
    ,'BONUS'                                                                       
    ,'PCTAFF'                                                                      
    ,'MEAL'                                                                        
    ,'OPEN'                                                                        
    ,'HRSOPEN'                                                                     
    ,'PSODA'                                                                       
    ,'PFRY'                                                                        
    ,'PENTREE'                                                                     
    ,'NREGS'                                                                       
    ,'NREGS11'                                                                     
    ,'TYPE2'                                                                       
    ,'STATUS2'                                                                     
    ,'DATE2'                                                                       
    ,'NCALLS2'                                                                     
    ,'EMPFT2'                                                                      
    ,'EMPPT2'                                                                      
    ,'NMGRS2'                                                                      
    ,'WAGE_ST2'                                                                    
    ,'INCTIME2'                                                                    
    ,'FIRSTIN2'                                                                    
    ,'SPECIAL2'                                                                    
    ,'MEALS2'                                                                      
    ,'OPEN2R'                                                                      
    ,'HRSOPEN2'                                                                    
    ,'PSODA2'                                                                      
    ,'PFRY2'                                                                       
    ,'PENTREE2'                                                                    
    ,'NREGS2'                                                                      
    ,'NREGS112']
dat = dat.replace('.', '')
dat = dat.apply(pd.to_numeric)
dat['EMPTOT']=dat['EMPPT']*0.5 + dat['EMPFT'] + dat['NMGRS']                                              
dat['EMPTOT2']=dat['EMPPT2']*0.5 + dat['EMPFT2'] + dat['NMGRS2']
df = dat[['STATE','EMPTOT','EMPTOT2']]

# create panel dataset for regression
df_1 = df[['STATE', 'EMPTOT']]
df_1['t']=0
df_2 = df[['STATE', 'EMPTOT2']]
df_2['t']=1
df_2.columns = ['STATE', 'EMPTOT','t']
df_reg = pd.concat([df_1, df_2], axis=0)
df_reg['dt'] = np.where((df_reg['STATE']== 1) & (df_reg['t']== 1) , 1, 0)

# regression
from statsmodels.formula.api import ols
ols = ols('EMPTOT ~ STATE + t + dt', data=df_reg).fit(cov_type='HC1',)
print(ols.summary())
```
